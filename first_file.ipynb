```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"festival Sales Data.csv",encoding="unicode_escape")
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
```

```python
#check if null value
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
#shows the total columns and rows
df.shape
```

```python
df.info()
```

```python
#change the data type
df["Amount"]=df["Amount"].astype("int")
```

```python
df["Amount"].dtypes
```

```python
df.columns
```

```python
#rename the column name
df.rename(columns={"Zone":"region"},inplace=True)
```

```python
df.info()
```

```python
#provides summary statistics of numerical columns
df[["Age","Amount","Orders"]].describe()
```

```python
#EXPLORATORY DATA ANALYSIS
sns.countplot(x="Gender",data=df,color="orange")
plt.show()
```

```python
ax=sns.countplot(x="Gender",data=df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
#sales by gender
sales_by_gender=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
ax=sns.barplot(x="Gender",y="Amount",data=sales_by_gender,color="pink")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#from the above graph we caan see most of buyers are females and even the purchasing powers are females are greater than male
```

```python
df.columns
```

```python
#sales by age
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#total sales by age
sale_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
colors=("orange","red","blue","pink","purple","green","brown")
ax=sns.barplot(x="Age Group",y="Amount",data=sale_age,hue="Age Group",palette=colors)
```

```python
#from above graph we can see that most of the buyers are of age group between 25-35 yrs female 
```

```python
#TOTAL NUMBER OF ORDERS FROM TOP 10 STATES
order_by_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
ax=sns.barplot(x="State",y="Orders",data=order_by_state)
```

```python
#TOTAL NUMBER OF sales FROM TOP 10 STATES
sales_by_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(x="State",y="Amount",data=sales_by_state)
```

```python
#FROM ABOVE GRAPH WE CAN SEE THAT MOST OF THE ORDERS AND TOTAL SALES FROM UP ,MAHARASHTRA AND KARNATAKA RESPECTIVELY
```

```python
#FROM THE ABOVE GRAPH WE CAN SEE MOST OF BUYERS ARE WORKING IN IT SECTOR,HEALTHCARE AND AVIATION. 
```

```python
#OCCUPATION
colors=("orange","pink", "gray")
sns.set(rc={"figure.figsize":(20,5)})
ax=sns.countplot(data=df,x="Occupation",palette=colors)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#FRIM THE ABOVE GRAM WE CAN SEE MOST OF BUYERS ARE FROM IT ,AVIATION AND  HEALTHCARE SECTOR

```
